```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "Qwen/Qwen2.5-0.5B-Instruct"

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
# Tokenize some stuff! 
# Show attention masks, decoding, padding

texts = [
    "The catcher sat on the mat.",
    "Tokenization is not the same as words."
]

# tokenizer("incredible")
# tokenizer.convert_ids_to_tokens([258, 57941])

input_ids = tokenizer(texts, return_tensors='pt', padding=True)

# model(input_ids=input_ids['input_ids'], attention_mask=input_ids['attention_mask'])
with torch.no_grad():
    out = model(**input_ids)
```

```python
# Try to apply chat template! 

input = tokenizer.apply_chat_template([{'role': 'user', 'content': 'Zaoshang Hao Zhongguo'}], add_generation_prompt=True, tokenize=True, return_tensors='pt')

with torch.no_grad():
    out = model.generate(**input)

print(tokenizer.decode(out[0]))
```

# Forward pass on a batch - predict the next token

```python
# Get model to run on a batch - inspect all the outputs

texts = [
    "The catcher sat on the mat.",
    "Tokenization is not the same as words."
]

input_ids = tokenizer(texts, return_tensors='pt', padding=True)
with torch.no_grad():
    out = model.generate(**texts)

print(tokenizer.decode(out[0]))
```

```python
# change attention to eager
model.config.use_cache = False
model.config.attn_implementation = "eager"  # โ FIRST
model.config.output_attentions = True       # โ THEN

output = model.generate(**input_ids)
print(output.attentions[0].shape)
```
