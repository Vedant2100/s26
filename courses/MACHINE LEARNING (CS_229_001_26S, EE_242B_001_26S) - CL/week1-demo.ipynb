```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "Qwen/Qwen2.5-0.5B-Instruct"

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
# Tokenize some stuff! 
# Show attention masks, decoding, padding

texts = [
    "The catcher sat on the mat.",
    "Tokenization is not the same as words."
]

# tokenizer("incredible")
# tokenizer.convert_ids_to_tokens([258, 57941])

input_ids = tokenizer(texts, return_tensors='pt', padding=True)

# model(input_ids=input_ids['input_ids'], attention_mask=input_ids['attention_mask'])
with torch.no_grad():
    out = model(**input_ids)
```

```python
# GENERATE method
# First, get the prefix
# Reminder: do_sample, max_new_tokens

inputs = tokenizer.apply_chat_template([{"role": "user", 
                                         "content": "Answer with Answer: <integer> only. Solve 1+1 =" }], 
                                       add_generation_prompt=True, 
                                       tokenize=True, return_tensors='pt')

with torch.no_grad():
    out = model.generate(inputs, max_new_tokens=100, temperature=100.)
```
