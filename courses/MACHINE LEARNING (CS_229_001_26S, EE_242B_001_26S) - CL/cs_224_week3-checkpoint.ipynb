### Matrices

We'll look at matrix inversion, decomposition, and numerical precision

```python
# Load and inspect

M = torch.load('XTX.pt')  # X_train^T @ X_train Loaded from HW 1

print('Shape: ', M.shape)
M[:5, :5]  
# Look - it's symmetric! *should* use operations for hermitian matrices
```

Let's look at inv and pinv. Spoiler: not even close! 

```python
torch.linalg.inv(M)[:5, :5]
```

Let's look at the "decomposition" of this matrix. 
We can use SVD, but because it is square symmetric, we could get the same result with better precision from using the specialized eigenvalue decomposition, eigh.

```python
U, S, Vt = torch.linalg.svd(M)
```

```python
# Look at the orthonormal matrices... 
```

```python
U @ Vt
```

```python
# Look at the singular values
S[:3]
S[-3:]
```

```python
# Is it ill conditioned? Condition number is max eigvalue over min 
condition = S[0] / S[-1]
print('condition? ', condition, ' = very bad')
```

```python
# Is this decomposition exactly the same as M? 
(M - U @ torch.diag(S) @ Vt).abs().max()  # max error entry
```

```python
# Compare the eigenvalue decomposition
eigvals, eigvecs = torch.linalg.eigh(M)
```

```python
(M - eigvecs @ torch.diag(eigvals) @ eigvecs.T).abs().max()
# Error using special hermitian/symmetric operators is lower
```

## K-Fold cross validation
And introducing sklearn, a library with tons of nice ML utilities

```python
# pip install scikit-learn
import sklearn  # Nice library with lots of ML tools
from sklearn.model_selection import KFold
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Build some synthetic data
X, y = make_classification(n_samples=20, n_features=2, n_informative=2, n_redundant=0, n_repeated=0, n_clusters_per_class=1)
```

```python
# Let's look at the data
plt.scatter(X[y==0,0], X[y==0,1], label='y=0')
plt.scatter(X[y==1,0], X[y==1,1], label='y=1')
plt.legend()
```

```python
kf = KFold(n_splits=5, shuffle=True)  # A KFold object
```

```python
print('Total number of points = ', len(X))

for train_idx, val_idx in kf.split(X):
    print(train_idx, val_idx)
```

```python
# Let's train a classifier, then do it with KFolds CV
clf = LogisticRegression(C=1.)  # init classifier
```

```python
clf.fit(X, y) # that was easy
```

```python
predicted_labels = clf.predict(X)
```

```python
accuracy_score(y, predicted_labels)
```
